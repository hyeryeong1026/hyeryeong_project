```python
import pandas as pd
cnts = pd.read_csv('read-counts.txt', sep='\t', comment='#', index_col=0)
cnts.head()
```

```python
cnts['clip_enrichment'] = cnts['CLIP-35L33G.bam'] / cnts['RNA-control.bam']
cnts['rden_change'] = (cnts['RPF-siLin28a.bam'] / cnts['RNA-siLin28a.bam']) / (cnts['RPF-siLuc.bam'] / cnts['RNA-siLuc.bam'])
cnts.head()
```

```python
from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.scatter(np.log2(cnts['clip_enrichment']),
           np.log2(cnts['rden_change']))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# pseudocount
pc = 1

# log2 enrichment 계산
cnts['CLIP_enrichment'] = np.log2((cnts['CLIP-35L33G.bam'] + pc) / (cnts['RNA-control.bam'] + pc))
cnts['RPF_siLin28a_density'] = np.log2((cnts['RPF-siLin28a.bam'] + pc) / (cnts['RNA-siLin28a.bam'] + pc))
cnts['RPF_siLuc_density'] = np.log2((cnts['RPF-siLuc.bam'] + pc) / (cnts['RNA-siLuc.bam'] + pc))
cnts['rden_change'] = cnts['RPF_siLin28a_density'] - cnts['RPF_siLuc_density']

# 필터링
mask = (
    (cnts['CLIP-35L33G.bam'] >= 30) &
    (cnts['RNA-control.bam'] >= 30) &
    (cnts['RNA-siLin28a.bam'] >= 30) &
    (cnts['RNA-siLuc.bam'] >= 30) &
    (cnts['RPF-siLin28a.bam'] >= 80) &
    (cnts['RPF-siLuc.bam'] >= 80)
)
filtered = cnts[mask].copy()

# NaN, inf 제거
x = filtered['CLIP_enrichment'].replace([np.inf, -np.inf], np.nan)
y = filtered['rden_change'].replace([np.inf, -np.inf], np.nan)
valid = x.notnull() & y.notnull()
x = x[valid]
y = y[valid]

# 상관계수
r, _ = pearsonr(x, y)

# 시각화
plt.figure(figsize=(5, 5), dpi=120)
plt.scatter(x, y, s=4, color='black', alpha=0.15)

# ⬇️ 축 범위: 실제 값 기준으로 여유 있게 설정
plt.xlim(x.min() - 0.5, x.max() + 0.5)
plt.ylim(y.min() - 0.2, y.max() + 0.2)

# 레이블 및 텍스트
plt.xlabel('LIN28A CLIP enrichment (log$_2$)')
plt.ylabel('Ribosome density change\nupon Lin28a knockdown (log$_2$)')
plt.xticks(range(int(np.floor(x.min())), int(np.ceil(x.max())) + 1, 2))
plt.yticks(range(int(np.floor(y.min())), int(np.ceil(y.max())) + 1, 1))
plt.grid(True, linestyle='-', linewidth=0.5, alpha=0.3)
plt.text(x.max() - 1.5, y.min() + 0.2, r'$r = %.4f$' % r, style='italic')
plt.title(r'$\bf{D}$  CLIP and ribosome footprinting upon Lin28a knockdown', loc='left')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import re
import math

# Pileup 파일 불러오기
pileup = pd.read_csv('CLIP-let7g-gene.pileup', sep='\t', names=['chrom', 'pos', '_ref', 'count', 'basereads', 'quals'])

# 불필요 기호 제거
toremove = re.compile(r'[<>*$#^]')
pileup['clean_bases'] = pileup['basereads'].apply(lambda x: toremove.sub('', x))

# 각 염기(A,T,C,G)의 개수 세기
def count_bases(s):
    s = s.upper()
    counts = {'A':0, 'T':0, 'C':0, 'G':0}
    for c in s:
        if c == '.' or c == ',':
            counts['match'] = counts.get('match', 0) + 1  # reference와 일치
        elif c in counts:
            counts[c] += 1
    return counts

counts_df = pd.DataFrame(pileup['clean_bases'].apply(count_bases).tolist())
counts_df['pos'] = pileup['pos']

# Shannon entropy 계산 함수
def shannon(row):
    total = sum([row[b] for b in 'ATCG'])
    if total == 0:
        return 0
    entropy = 0
    for base in 'ATCG':
        p = row[base] / total
        if p > 0:
            entropy -= p * math.log2(p)
    return entropy

counts_df['entropy'] = counts_df.apply(shannon, axis=1)
```

```python
with open('CLIP-let7g.entropy.bedgraph', 'w') as f:
    f.write('track type=bedGraph name="Entropy let7g" description="Shannon entropy from CLIP"\n')
    for _, row in counts_df.iterrows():
        chrom = 'chr9'
        start = int(row['pos']) - 1
        end = int(row['pos'])
        value = row['entropy']
        f.write(f'{chrom}\t{start}\t{end}\t{value:.4f}\n')
```

```python
import pandas as pd
from collections import defaultdict, Counter
import matplotlib.pyplot as plt

def extract_cims_positions(pileup_file):
    """pileup 파일에서 CIMS 위치 추출"""
    cims_positions = []

    with open(pileup_file) as f:
        for line in f:
            fields = line.strip().split('\t')
            chrom = fields[0]
            pos = int(fields[1])
            ref_base = fields[2]
            reads = fields[4]

            # Deletion이나 substitution 검색
            mutations = 0
            total_reads = len([c for c in reads if c not in ['^', '$']])

            # Deletion (*) 카운트
            deletions = reads.count('*')

            # Substitution 카운트 (대문자 A,T,C,G)
            substitutions = sum(1 for c in reads if c in 'ATCG')

            mutations = deletions + substitutions

            # 5% 이상 mutation rate일 때 CIMS로 간주
            if total_reads > 10 and mutations / total_reads > 0.05:
                cims_positions.append({
                    'chrom': chrom,
                    'pos': pos,
                    'ref': ref_base,
                    'mut_rate': mutations / total_reads,
                    'mut_type': 'del' if deletions > substitutions else 'sub'
                })

    return pd.DataFrame(cims_positions)

# 각 let-7 유전자별 CIMS 추출
cims_let7g = extract_cims_positions('CLIP-let7g-gene.pileup')
cims_let7f1 = extract_cims_positions('CLIP-let7f1-gene.pileup')
cims_let7d = extract_cims_positions('CLIP-let7d-gene.pileup')

print(f"let7g CIMS sites: {len(cims_let7g)}")
print(f"let7f1 CIMS sites: {len(cims_let7f1)}")
print(f"let7d CIMS sites: {len(cims_let7d)}")
```

```python
import pandas as pd
from collections import defaultdict, Counter
import matplotlib.pyplot as plt

# 1. CIMS 위치 추출
def extract_cims_positions(pileup_file):
    """pileup 파일에서 CIMS 위치 추출"""
    cims_positions = []

    with open(pileup_file) as f:
        for line in f:
            fields = line.strip().split('\t')
            chrom = fields[0]
            pos = int(fields[1])
            ref_base = fields[2]
            reads = fields[4]

            # Deletion이나 substitution 검색
            mutations = 0
            total_reads = len([c for c in reads if c not in ['^', '$']])

            deletions = reads.count('*')
            substitutions = sum(1 for c in reads if c in 'ATCG')
            mutations = deletions + substitutions

            # 5% 이상 mutation rate일 때 CIMS로 간주
            if total_reads > 10 and mutations / total_reads > 0.05:
                cims_positions.append({
                    'chrom': chrom,
                    'pos': pos,
                    'ref': ref_base,
                    'mut_rate': mutations / total_reads,
                    'mut_type': 'del' if deletions > substitutions else 'sub'
                })

    return pd.DataFrame(cims_positions)

# 2. CIMS 추출 실행
cims_let7g = extract_cims_positions('CLIP-let7g-gene.pileup')
print(f"let7g CIMS sites: {len(cims_let7g)}")

# 3. 서열 추출 함수 정의
def extract_flanking_sequences(cims_df, reference_seq, window=10):
    """CIMS 위치 주변 서열 추출"""
    flanking_seqs = []

    for idx, row in cims_df.iterrows():
        pos = row['pos'] - 1  # 0-based indexing
        start = max(0, pos - window)
        end = min(len(reference_seq), pos + window + 1)

        if end - start >= window * 2:
            seq = reference_seq[start:end].upper()
            flanking_seqs.append({
                'pos': row['pos'],
                'sequence': seq,
                'cims_pos': window,
                'mut_rate': row['mut_rate']
            })

    return pd.DataFrame(flanking_seqs)

# 4. 참조 서열 정의 (예시 - 실제로는 fasta 파일에서 추출)
let7g_seq = "UGAGGUAGUAGUUUGUACAGUUUGGGGAGCCUGGUUGACUAUCAUGAAGGCUAAGAUUCUGGGGAGGUAGUAGGUUGUACAGUUGAAACUCGGCCUAAGGAGCCUGGGUUGACUGUAGACAGAAGGCUGAA"

# 5. **중요: flanking_seqs 변수 생성**
flanking_seqs = extract_flanking_sequences(cims_let7g, let7g_seq)
print(f"Flanking sequences extracted: {len(flanking_seqs)}")

# 6. 이제 motif 분석 실행 가능
def analyze_motifs(flanking_df, motif_length=4):
    """k-mer motif 빈도 분석"""
    motif_counts = Counter()
    total_windows = 0

    for idx, row in flanking_df.iterrows():
        seq = row['sequence']
        cims_pos = row['cims_pos']

        for i in range(max(0, cims_pos-5), min(len(seq)-motif_length+1, cims_pos+6)):
            motif = seq[i:i+motif_length]
            if len(motif) == motif_length and 'N' not in motif:
                motif_counts[motif] += 1
                total_windows += 1

    motif_freq = {motif: count/total_windows for motif, count in motif_counts.items()}
    return motif_freq

# 7. 분석 실행
if len(flanking_seqs) > 0:
    motif_freq = analyze_motifs(flanking_seqs)

    # 상위 motif 시각화
    top_motifs = sorted(motif_freq.items(), key=lambda x: x[1], reverse=True)[:20]

    plt.figure(figsize=(12, 6))
    motifs, freqs = zip(*top_motifs)
    bars = plt.bar(range(len(motifs)), freqs, color='steelblue', alpha=0.7)

    # GGAG 강조
    for i, motif in enumerate(motifs):
        if 'GGAG' in motif:
            bars[i].set_color('red')
            bars[i].set_alpha(0.9)

    plt.xlabel('4-mer Motifs')
    plt.ylabel('Frequency')
    plt.title('CIMS Flanking Region Motif Enrichment')
    plt.xticks(range(len(motifs)), motifs, rotation=45)
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()

    print(f"GGAG motif frequency: {motif_freq.get('GGAG', 0):.3f}")
else:
    print("No CIMS sites found or no flanking sequences extracted")
```

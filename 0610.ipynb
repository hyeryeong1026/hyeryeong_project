```python
import pandas as pd
import re
import math

# Pileup 파일 불러오기
pileup = pd.read_csv('CLIP-let7g-gene.pileup', sep='\t', names=['chrom', 'pos', '_ref', 'count', 'basereads', 'quals'])

# 불필요 기호 제거
toremove = re.compile(r'[<>*$#^]')
pileup['clean_bases'] = pileup['basereads'].apply(lambda x: toremove.sub('', x))

# 각 염기(A,T,C,G)의 개수 세기
def count_bases(s):
    s = s.upper()
    counts = {'A':0, 'T':0, 'C':0, 'G':0}
    for c in s:
        if c == '.' or c == ',':
            counts['match'] = counts.get('match', 0) + 1  # reference와 일치
        elif c in counts:
            counts[c] += 1
    return counts

counts_df = pd.DataFrame(pileup['clean_bases'].apply(count_bases).tolist())
counts_df['pos'] = pileup['pos']

# Shannon entropy 계산 함수
def shannon(row):
    total = sum([row[b] for b in 'ATCG'])
    if total == 0:
        return 0
    entropy = 0
    for base in 'ATCG':
        p = row[base] / total
        if p > 0:
            entropy -= p * math.log2(p)
    return entropy

counts_df['entropy'] = counts_df.apply(shannon, axis=1)
```

```python
with open('CLIP-let7g.entropy.bedgraph', 'w') as f:
    f.write('track type=bedGraph name="Entropy let7g" description="Shannon entropy from CLIP"\n')
    for _, row in counts_df.iterrows():
        chrom = 'chr9'
        start = int(row['pos']) - 1
        end = int(row['pos'])
        value = row['entropy']
        f.write(f'{chrom}\t{start}\t{end}\t{value:.4f}\n')
```

```python
import pandas as pd

def parse_junctions(sam_file):
    junctions = []
    with open(sam_file) as f:
        for line in f:
            parts = line.strip().split('\t')
            chrom = parts[2]
            start = int(parts[3])
            cigar = parts[5]
            if 'N' in cigar:
                # N이 어디서 발생했는지 추정 (간단 버전)
                import re
                matches = re.findall(r'(\d+)([MIDNSHP=X])', cigar)
                pos = start
                for length, op in matches:
                    length = int(length)
                    if op == 'N':
                        j_start = pos
                        j_end = pos + length
                        junctions.append((chrom, j_start, j_end))
                    if op in 'MDN=X':
                        pos += length
    return pd.DataFrame(junctions, columns=['chrom', 'start', 'end'])

lin28a_junc = parse_junctions("lin28a_junction.sam")
luc_junc = parse_junctions("luc_junction.sam")
```

```python
# junction별 count 정리
lin28a_counts = lin28a_junc.value_counts().reset_index(name='lin28a_count')
luc_counts = luc_junc.value_counts().reset_index(name='luc_count')

# merge
merged = pd.merge(lin28a_counts, luc_counts, on=['chrom', 'start', 'end'], how='outer').fillna(0)

# splicing 차이 큰 것 추출
merged['delta'] = merged['lin28a_count'] - merged['luc_count']
merged['abs_delta'] = merged['delta'].abs()
merged.sort_values('abs_delta', ascending=False).head()
```

```python
def classify_event(start, end):
    length = end - start
    if length > 1000:
        return "exon_skipping"
    elif 30 < length < 300:
        return "intron_retention"
    else:
        return "other"

# 기존 merged DataFrame 사용
merged['event_type'] = merged.apply(lambda row: classify_event(row.start, row.end), axis=1)

# 분포 보기
print(merged['event_type'].value_counts())

# 시각화
import matplotlib.pyplot as plt
merged['event_type'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Splicing Event Type Distribution")
plt.xlabel("Event Type")
plt.ylabel("Count")
plt.show()
```

```python
# 변화량 큰 junction 100개 선택
top_junc = merged.sort_values('abs_delta', ascending=False).head(100)

# BED로 저장 (start 부근 ±30nt)
with open("junction_flanks.bed", "w") as f:
    for i, row in top_junc.iterrows():
        flank_start = max(0, int(row.start) - 30)  # 0보다 작으면 안 됨
        flank_end = int(row.start) + 30
        f.write(f"{row.chrom}\t{flank_start}\t{flank_end}\tjunc{i}\n")
```

```python
# GTF 파일에서 gene_id ↔ gene_name 추출
gtf_file = "/content/drive/MyDrive/binfo1-datapack1/gencode.gtf"

import pandas as pd
import re

mapping = {}
with open(gtf_file) as f:
    for line in f:
        if line.startswith("#"):
            continue
        fields = line.strip().split('\t')
        if fields[2] != 'gene':
            continue
        attrs = fields[8]
        gid_match = re.search('gene_id "([^"]+)"', attrs)
        gname_match = re.search('gene_name "([^"]+)"', attrs)
        if gid_match and gname_match:
            gid = gid_match.group(1).split('.')[0]  # remove .2 버전
            gname = gname_match.group(1)
            mapping[gid] = gname

# 매핑 결과를 DataFrame으로 변환
gene_map = pd.DataFrame.from_dict(mapping, orient='index', columns=['symbol'])
gene_map.index.name = 'gene_id'
gene_map.head()
```

```python
# cnts의 index는 ENSMUSG00000000001.1 형태니까 . 제거하고 . 앞부분만 추출
cnts.index.name = 'gene_id'
cnts.index = cnts.index.str.replace(r'\.\d+', '', regex=True)

# symbol 붙이기
cnts = cnts.merge(gene_map, how='left', left_index=True, right_index=True)

# 확인
cnts[['symbol']].head()
```

```python
# symbol 기준으로 필터링
rbps = [
    "Hnrnpa1", "Hnrnpa2b1", "Hnrnpc", "Hnrnpd", "Hnrnph1", "Hnrnph2", "Hnrnph3",
    "Hnrnpl", "Hnrnpm", "Hnrnpk", "Hnrnpr", "Hnrnpab", "Hnrnpf", "Hnrnpu",
    "Srsf1", "Srsf2", "Srsf3", "Srsf4", "Srsf5", "Srsf6", "Srsf7", "Srsf9", "Srsf10",
    "Rbm5", "Rbm24", "Rbm38", "Rbm45", "Ptbp1", "Ptbp2",
    "Tia1", "Tial1", "Fus", "Ewsr1", "Tra2a", "Tra2b"
]

rbp_counts = cnts[cnts['symbol'].isin(rbps)]

# rden_change 계산 다시
eps = 1e-3
cnts['rden_change'] = (
    (cnts['RPF-siLin28a.bam'] + eps) / (cnts['RNA-siLin28a.bam'] + eps)
) / (
    (cnts['RPF-siLuc.bam'] + eps) / (cnts['RNA-siLuc.bam'] + eps)
)

# RBP subset에서도 다시 붙이기
rbp_counts['rden_change'] = cnts.loc[rbp_counts.index, 'rden_change']
rbp_counts = rbp_counts.set_index('symbol')
rbp_counts = rbp_counts.sort_values('rden_change')

# 시각화
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# 검정색 막대
plt.barh(
    rbp_counts.index,
    rbp_counts['rden_change'],
    color='black'
)

# 강조된 기준선
plt.axvline(1.0, color='gray', linestyle='--', linewidth=2)
plt.text(1.01, -1, 'No Change', color='gray', fontsize=10)

# 라벨 및 제목
plt.xlabel("Translation Efficiency Change (rden_change)", fontsize=12)
plt.title("RBP Translation Efficiency after LIN28A Knockdown", fontsize=14, pad=15)

# y축 tick label 크기 조정
plt.tick_params(axis='y', labelsize=10)
plt.grid(axis='x', linestyle=':', alpha=0.3)

plt.tight_layout()
plt.show()

```
